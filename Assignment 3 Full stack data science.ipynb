13. What can you do to save a programme from crashing if it encounters an error?

```python
 # we can handle the error by using Try except block to catch the errors without crashing the programme
```

```python
try:
    a=1/0
    print(a)
except Exception as e:
    print("The error is",e) #This will not crash the programme
print("This will not crash the programme")
```
