10. How can you tell the difference between break and continue?

```python
break will make destroy the whole upcoming code in the loop and jumps to next block
```

```python
for i in range(10):
    if i==5:
        break
    else:
        print(i)
print("end")
```

```python
continue skips only that instance and loop other values as it is
```

```python
for i in range(10):
    if i==5:
        continue
    else:
        print(i)
print("End")
```
